```python
data = df[df.columns[2:-1]]
target = df[df.columns[-1]]
```

## test case split

```python
from sklearn.model_selection import train_test_split
```

```python
xtr, xte, ytr, yte = train_test_split(data, target, test_size=0.3)
```

## without grid_search

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xtr, ytr)
```

```python
pred_xte = rfc.predict(xte)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acscore = accuracy_score
```

```python
print(acscore(pred_xte, yte))
```

## grid search

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'max_depth': [3, 5, 6, 7, 10], 'min_samples_split': [1, 2, 3,]}
```

```python
grid_forest = GridSearchCV(rfc, param_grid = parameters, cv=5, refit=True,)
```

```python
grid_forest.fit(xtr, ytr)
```

```python
score_df = pd.DataFrame(grid_forest.cv_results_)
```

```python
score_df[['params', 'mean_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score', 'split3_test_score', 'split4_test_score']].sort_values('mean_test_score')
```

```python
score_df
```

```python
grid_forest.best_params_
```

```python
grid_forest.best_score_
```

```python
rfc = RandomForestClassifier(max_depth=10, min_samples_split=2)
```

```python
rfc.fit(xtr, ytr)
```

```python
predicts = rfc.predict(xte)
```

```python
print(acscore(predicts, yte))
```
